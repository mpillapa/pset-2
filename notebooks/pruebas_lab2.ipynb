```python
import pandas as pd


year_start = 2015
year_end   = 2016
months     = [1]  

base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-{month:02d}.parquet"

frames = []
for year in range(year_start, year_end + 1):
    for month in months:
        url = base_url.format(year=year, month=month)
        pd_temp = pd.read_parquet(url)
        pd_temp.columns = pd_temp.columns.str.lower()
        frames.append(pd_temp)

df = pd.concat(frames, ignore_index=True)
df.head()
```

```python
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-10.parquet"

df_2015 = pd.read_parquet(base_url.format(year=2015))
df_2025 = pd.read_parquet(base_url.format(year=2025))

df_2015.columns = df_2015.columns.str.lower()
df_2025.columns = df_2025.columns.str.lower()


cols_2015 = set(df_2015.columns)
cols_2025 = set(df_2025.columns)

print("Columnas 2015:", sorted(df_2015.columns.tolist()))
print()
print("Columnas 2025:", sorted(df_2025.columns.tolist()))
print()
print("Solo en 2015:", sorted(cols_2015 - cols_2025))
print("Solo en 2025:", sorted(cols_2025 - cols_2015))
print("En ambas:    ", sorted(cols_2015 & cols_2025))
```
