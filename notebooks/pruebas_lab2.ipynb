```python
import pandas as pd


year_start = 2015
year_end   = 2016
months     = [1]  

base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-{month:02d}.parquet"

frames = []
for year in range(year_start, year_end + 1):
    for month in months:
        url = base_url.format(year=year, month=month)
        pd_temp = pd.read_parquet(url)
        pd_temp.columns = pd_temp.columns.str.lower()
        frames.append(pd_temp)

df = pd.concat(frames, ignore_index=True)
df.head()
```

```python
import pandas as pd


year_start = 2015
year_end   = 2025
months     = [1]  

base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{year}-{month:02d}.parquet"

frames = []
for year in range(year_start, year_end + 1):
    for month in months:
        url = base_url.format(year=year, month=month)
        pd_temp = pd.read_parquet(url)
        pd_temp.columns = pd_temp.columns.str.lower()
        frames.append(pd_temp)

one_month_year = pd.concat(frames, ignore_index=True)
one_month_year.head()
```

```python
one_month_year.describe()
```

```python
# Valindando que años contiene la columna tpep_pickup_datetime
one_month_year['tpep_pickup_datetime'] = pd.to_datetime(one_month_year['tpep_pickup_datetime'])
one_month_year['year'] = one_month_year['tpep_pickup_datetime'].dt.year
one_month_year['year'].value_counts().sort_index()
```

```python
# verificamos la cantidad de nulos por columna
one_month_year.isna().sum()
```

```python
# verificamos la cantidad de nulos por columna agrupados por año
nulos_por_columna_año = one_month_year.groupby('year').apply(lambda x: x.isna().sum())
nulos_por_columna_año
```
